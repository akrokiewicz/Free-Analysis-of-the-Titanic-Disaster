### Titanic dataset info:
Popular "Titanic" dataset often used in analytics trainings.

Original loaded data contains 1309 rows (covering only passengers, not the crew) and 14 columns. Columns meaning:
- pclass - ticket class
    * 1st = Upper
    * 2nd = Middle
    * 3rd = Lower
- survived - determins if person survived
- name - passenger name
- sex - sex
- age - age is fractional if less than 1. If the age is estimated, is it in the form of xx.5
- sibsp - number of siblings/spouses aboard
- parch - number of parents/children aboard. Some children travelled only with a nanny, therefore parch is 0 for them.
- ticket - ticket number
- fare - passenger fare
- cabin - cabin number
- embarked - port of embarkation
    * C = Cherbourg 
    * Q = Queenstown
    * S = Southampton
- boat - lifeboat (if survived)
- body - body number (if did not survive and body was recovered)
- home.dest

New columns added during the processing:
- name_prefix (object) - cut name prefix from name column
- age_range (object)- age range, split every 10 years
- class_num (integer) - numeric assignment for class:
    * 1 = 1st
    * 2 = 2nd
    * 3 = 3rd
- survived_num - numeric assignment for survival trigger:
    * 0 = dead 
    * 1 = survived
- sex_num - numeric assignment for sex:
    * 1 - female
    * 2 - male
- family_size_aboard:
    * 1 = when alone
    * value > 1 = sum of sibsp and parch plus 1

### Cleaning dataset, preparing to analysys - data mining

```python
#checking null values in order to decide how to proceed
data.isnull().sum()
```

#### Missing values:

```python
msno.matrix(data, figsize=(12,7)) #visualisation of missing values
```

```python
#looking for data unable to visualize (text values mamong numeric values)
data['age'].value_counts()
```

```python
#dataset is too small to get rid all "unknown" values from such important variable as age. 
#considering that prefix from full name may be connected to age, 
#name_prefix will divide data into prefix groups and age will be filled with a mean value for a group
data['name_prefix'] = data['name'].apply(lambda x: x[(x.find(",")+1):(x.find(".")+1)].strip())
data['name_prefix'].value_counts()
```

```python
#defined in case if needed for more columns 
def unknown_to_nan(row):
    """
    Fuction takes a data row from specific column as an argument and returns NaN value when finds "unknown". 
    Returns an original value for any other condition.
    Input: data frame column
    Output: data frame column without values egual to "unknown"
    """
    return (np.NaN) if row == "unknown" else row
```

```python
#changing "age" to numeric value
data['age'] = data['age'].apply(unknown_to_nan)
data['age'] = data['age'].astype("float")

#populating missing values with mean age value for created prefix
data["age"] = data.groupby("name_prefix")['age'].transform(lambda x: x.fillna(x.mean()))

#remove unnecessary column
data.drop(columns = 'name_prefix', inplace=True)
```

```python
#control check of age values
data['age'].value_counts()
```

```python
data.info(memory_usage='deep')
```

```python
#replacing NaN values with 0s
data['sibsp'].fillna(0, inplace=True)
data['parch'].fillna(0, inplace=True)

#and changing them to numeric
data['sibsp'] = data['sibsp'].astype("int")
data['parch'] = data['parch'].astype("int")
```

```python
data['pclass'].value_counts()
```

```python
data['sex'].value_counts()
```

```python
data['survived'].value_counts()
```

```python
data['embarked'].value_counts()
```

```python
#changing columns type in order to lower memoory usage
data['pclass'] = data['pclass'].astype('category')
data['sex'] = data['sex'].astype('category')
data['survived'] = data['survived'].astype('category')
data['embarked'] = data['embarked'].astype('category')
```

```python
data.info(memory_usage='deep')
```

```python
def age_range(row):
    """
    Creates age ranges about each 10 years to group passengers by age. 
    Input: data frame column containing age.
    """
    if row <= 10:
        return '0-10'
    elif 10 < row <= 20:
        return '10-20'
    elif 20 < row <= 30:
        return '20-30'
    elif 30 < row <= 40:
        return '30-40'
    elif 40 < row <= 50:
        return '40-50'
    elif 50 < row <= 60:
        return '50-60'
    elif 60 < row <= 70:
        return '60-70'
    elif 70 < row:
        return '>70'

data['age_range'] = data['age'].apply(age_range)
```

```python
#Changing values into numbers in order to calculate correlation:
def class_num(row):
    """
    Changes text value of passenger class into numeric value.
    """
    if row == '1st':
        return 1
    elif row == '2nd':
        return 2
    elif row == '3rd':
        return 3
    
data['pclass_num'] = data['pclass'].apply(class_num).astype("int")
```

```python
data['survived_num'] = data['survived'].map({"Yes": 1, "No": 0}).astype('int')
```

```python
#creating family size to see influence for survival rate
data['family_size_aboard'] = (data['sibsp'] + data['parch'] + 1).astype('int')

#create column with 1 when body has been found and 0 when not (or person stayed alive)
data['body_found'] = data['body'].apply(lambda x: 1 if pd.notna(x) else 0)
data['body_found'].value_counts()
```

```python
#One hot-encoding for sex field
data = pd.concat([data, pd.get_dummies(data['sex'], prefix = 'sex', prefix_sep = "_", drop_first = False)], axis=1)
```

```python
data.info()
```

Split data for categories - survived and not_survived:

```python
#Creating filtered data survived and not_survived to have a bettter look into groups
survived = data[data['survived_num'] == 1]
not_survived = data[data['survived_num'] == 0]
```

# Titanic survival prediction

```python
#CREATING FEATURES AND PREDICTED VALUE:
#fields sibsp and parch will be skipped due to high correlation with field family_size_abroad which will be usead as one calculated output for two mentioned fields
#fare will be skipped due to high correlation with pclass field, age_range with age
#field name seems to be unrelevant, same as cabin, home.dest and embarked
#boad and body are variables without influence to the outcome. Those can can't be provided before the prediction.
features = ['pclass_num', 'sex_female', 'sex_male', 'age', 'family_size_aboard']

X = data[features]
y = data['survived_num']
```

```python
#Split into test and train data:
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=1)
```

```python
#Scaling values
sc = StandardScaler()
train_X = sc.fit_transform(train_X)
test_X = sc.transform(test_X)
```

## Finding a proper model
### Decision Tree Classifier

```python
survival_model = DecisionTreeClassifier(random_state=1)
```

```python
survival_model.fit(train_X, train_y)
```

```python
survival_predictions = survival_model.predict(test_X)
```

```python
mi_score = mutual_info_classif(X,y, random_state=1)
```

```python
zipped = zip(X.columns, mi_score)
```

```python
print("Correlation between variables and the outcome (MI score):")
for x, y in zipped:
    print("MI score for variable {} is {:0.2f}.".format(x, y))
```

```python
feature_importance = zip(X.columns ,survival_model.feature_importances_)
```

```python
for x, y in feature_importance:
    print("{} importance is {:0.2f}.".format(x, y))
```

```python
fig, ax = plt.subplots(figsize=(5,5))
plot_confusion_matrix(survival_model, test_X, test_y, ax=ax) 
plt.show()
```

```python
accuracy = accuracy_score(survival_predictions, test_y)
```

```python
print("Accuracy score is: {}.".format(accuracy))
```

```python
print(classification_report(test_y, survival_predictions))
```

Let's try to find a better model:

```python
for max_leaf_nodes in [5, 10, 15, 20, 25, 30, 50]:
    print("\n\n\nREASULT FOR {} MAX LEAF NODES:".format(max_leaf_nodes))
    survival_model = DecisionTreeClassifier(max_leaf_nodes=max_leaf_nodes, random_state=1)
    survival_model.fit(train_X, train_y)
    survival_predictions = survival_model.predict(test_X)
    
    fig, ax = plt.subplots(figsize=(5,5))
    plot_confusion_matrix(survival_model, test_X, test_y, ax=ax) 
    plt.show()
    
    print(classification_report(test_y, survival_predictions))
    
```

```python
for min_samples_leaf in [3, 5, 10, 20, 50]:
    print("\n\n\nREASULT FOR {} MIN SAMPLES LEAF:".format(min_samples_leaf))
    survival_model = DecisionTreeClassifier(max_leaf_nodes=25, min_samples_leaf=min_samples_leaf, random_state=1)
    survival_model.fit(train_X, train_y)
    survival_predictions = survival_model.predict(test_X)
    
    fig, ax = plt.subplots(figsize=(5,5))
    plot_confusion_matrix(survival_model, test_X, test_y, ax=ax) 
    plt.show()
    
    print(classification_report(test_y, survival_predictions))
    
```

```python
for max_depth in [5, 10, 15, 20, 25, 50, 100]:
    print("\n\n\nREASULT FOR {} MAX DEPTH:".format(max_depth))
    survival_model = DecisionTreeClassifier(max_leaf_nodes=25, min_samples_leaf=10, max_depth=max_depth, random_state=1)
    survival_model.fit(train_X, train_y)
    survival_predictions = survival_model.predict(test_X)
    
    fig, ax = plt.subplots(figsize=(5,5))
    plot_confusion_matrix(survival_model, test_X, test_y, ax=ax) 
    plt.show()
    
    print(classification_report(test_y, survival_predictions))
```

The winner in this category is below:

```python
survival_model = DecisionTreeClassifier(max_leaf_nodes=25, min_samples_leaf=10, max_depth=25, random_state=1)
survival_model.fit(train_X, train_y)
survival_predictions = survival_model.predict(test_X)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
plot_confusion_matrix(survival_model, test_X, test_y, ax=ax) 
plt.show()
    
print(classification_report(test_y, survival_predictions))
```
